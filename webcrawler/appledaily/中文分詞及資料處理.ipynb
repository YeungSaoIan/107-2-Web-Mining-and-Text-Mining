```python
#從wiki爬台灣政治人物
#台北市政治人物
# 對wiki API發出請求
import requests

url = "https://zh.wikipedia.org/w/api.php"
params = {
    "action":"query",
    "list":"search", 
    "format":"json",
    "srsearch":"incategory:台北市政治人物",
    "srlimit":"500",
    "srprop":"size"
}
res = requests.get(url, params=params)
nested = res.json()
print(nested)
```

```python
# 小試身手2 解答
names = []
for item in nested["query"]["search"]:
    names.append(item["title"] + "\n")

with open("userdict.txt", "w", encoding="utf-8") as O:
    for name in names:
        O.write(name + "\n")
```
