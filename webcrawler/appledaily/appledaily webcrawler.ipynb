```python
from datetime import datetime #讀取datetime套件
from time import sleep #讀取time套件
from random import random #讀取random套件
import json #讀取json套件
 
from requests import Session #讀取requests套件
from bs4 import BeautifulSoup #讀取BeautifulSoup套件
 
session = Session()
session.headers['User-Agent'] = 'Angel' #把瀏覽器從python client偽裝成其他名字，以免被系統誤判為惡意網路攻擊

base_url = 'https://tw.news.appledaily.com/international/realtime/{page}' #蘋果新聞網頁版連結
page = 1
links = []
current_date = datetime.now().date()
 
while True:
    url = base_url.format(page=page)
    response = session.get(url)
    print(f'已經爬到第{page}頁')
    dom = BeautifulSoup(response.text)
    raw_time = dom.select('h1.dddd > time')[0].text
    date = datetime.strptime(raw_time, '%Y / %m / %d').date()
    if date < current_date:
        break
    elements = dom.select('h1.dddd + ul.rtddd > li')
    for element in elements:
        link = element.select('a')[0]['href']
        links.append(link)
    sleep(random() * 5) #生成隨機爬蟲時間，以免被系統誤判為惡意網路攻擊
    page += 1
 
base_url = 'https://tw.appledaily.com/new/realtime/{page}'
page = 1
links = []
current_date = datetime.now().date()
 
while True:
    url = base_url.format(page=page)
    response = session.get(url)
    print(f'已經爬到第{page}頁面')
    dom = BeautifulSoup(response.text)
    raw_time = dom.select('h1.dddd > time')[0].text
    date = datetime.strptime(raw_time, '%Y / %m / %d').date()
    if date < current_date:
        break
    elements = dom.select('h1.dddd + ul.rtddd > li')
    for element in elements:
        link = element.select('a')[0]['href']
        links.append(link)
    sleep(random() * 5)
    page += 1
 
htmls = []
for num, link in enumerate(links):
    print(f'還剩下{len(links) - num}個頁面')
    response = session.get(link)
    htmls.append(response.text)
    sleep(random() * 5)
 
posts = []
for html in htmls:
    dom = BeautifulSoup(html)
    title = dom.select('article.ndArticle_leftColumn h1')[0].text #爬取新聞標題，目標抓取'article.ndArticle_leftColumn h1'標籤
    created_time = dom.select('article.ndArticle_leftColumn div.ndArticle_creat')[0].text #爬取新聞發佈時間
    category = dom.select('nav div.ndgTag a.current')[0].text #爬取新聞類別，目標抓取'nav div.ndgTag a.current'籤
    content = dom.select('article.ndArticle_content p')[0].text #爬取新聞內文，目標抓取'article.ndArticle_content p'標籤
    post = dict(title=title, created_time=created_time, category=category, content=content)
    posts.append(post)
 
with open('appledaily2.json', 'w') as f: #爬取新聞資料後以json格式另存新檔
    json.dump(posts, f, ensure_ascii=False, indent=2)
```
