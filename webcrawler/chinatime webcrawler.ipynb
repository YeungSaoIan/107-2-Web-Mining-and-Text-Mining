```python
import requests
from bs4 import BeautifulSoup
```

```python
t = requests.get("https://www.chinatimes.com/chinese/total?chdtv") #將此頁面的HTML GET下來
print(t.text) #印出HTML
```

```python
url = "https://www.chinatimes.com/chinese/total?page=1&chdtv"
newsrecord = []
for i in range(10):   
    t = requests.get(url)
    drink = BeautifulSoup(t.text,"html.parser")
    giveme = drink.select("h3.title a") #取<a>標籤存入giveme
    showmetime = drink.select("div.meta-info time")#取<meta-info>標籤存入showmetime
    #print ("本頁的URL為"+url)
    url = "https://www.chinatimes.com/chinese/total?page="+ str(i+2) +"&chdtv" #下一頁的網址
    for u in giveme:
        print(u["href"], u.text) 
        newsrecord += [u.text]
```

```python
url = "https://www.chinatimes.com/chinese/total?page=1&chdtv"
timerecord = []
for i in range(10):   
    t = requests.get(url)
    drink = BeautifulSoup(t.text,"html.parser")
    showmetime = drink.select("div.meta-info time")#取<meta-info>標籤存入showmetime
    #print ("本頁的URL為"+url)
    url = "https://www.chinatimes.com/chinese/total?page="+ str(i+2) +"&chdtv" #下一頁的網址
    for v in showmetime:
        print(v["datetime"]) 
        timerecord += [v["datetime"]]
```

```python
fullrecord = []
for i in range(len(newsrecord)):
    print(newsrecord[i],timerecord[i])
    full = []
    full = newsrecord[i],timerecord[i]
    print(full)
    fullrecord += full
```
