```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
t = requests.get("https://www.chinatimes.com/chinese/total?chdtv") #將此頁面的HTML GET下來
print(t.text) #印出HTML
```

```python
url = "https://www.chinatimes.com/chinese/total?page=1&chdtv"
records = []
posts = []
for i in range(10):   
    t = requests.get(url)
    drink = BeautifulSoup(t.text,"html.parser")
    title = drink.select("h3.title a") #取<a>標籤存入giveme
    time = drink.select("div.meta-info time")#取<meta-info>標籤存入showmetime
    #print ("本頁的URL為"+url)
    url = "https://www.chinatimes.com/chinese/total?page="+ str(i+2) +"&chdtv" #下一頁的網址
    for u in title:
        print(u["href"], u.text) 
        records += [u.text]
```

```python
url = "https://www.chinatimes.com/chinese/total?page=1&chdtv"
created_times = []
for i in range(10):   
    t = requests.get(url)
    drink = BeautifulSoup(t.text,"html.parser")
    time = drink.select("div.meta-info time")#取<meta-info>標籤存入showmetime
    #print ("本頁的URL為"+url)
    url = "https://www.chinatimes.com/chinese/total?page="+ str(i+2) +"&chdtv" #下一頁的網址
    for v in time:
        print(v["datetime"]) 
        created_times += [v["datetime"]]
```

```python
posts = []
for i in range(len(records)):
    title = records[i]
    created_time = created_times[i]
    post = dict(title=title, created_time=created_time)
    posts.append(post)
```

```python
with open('chinatime.json', 'w') as f:
    json.dump(posts, f, ensure_ascii=False, indent=2)
```
