```python
import csv
csvfile = open('testoutput.csv', 'r', encoding='utf-8')
reader = csv.reader(csvfile)
```

```python
type(reader)
```

```python
newsinput = []
for row in reader:
    #print(row)
    newsinput.append(row) 
```

```python
import numpy as np

ary = np.asarray(newsinput)
```

```python
ary = np.concatenate(ary)
ary
```

```python
f = open('cuted.txt', 'w', encoding='utf-8')
#cut = []
length = len(ary)
i = 0
for i in range(length):
    if i %2 == 0:
        #print(ary[i])
        seg_list = jieba.cut(ary[i])
#        cut = "  ".join(seg_list)
        f.write("  ".join(seg_list))
        f.write('\n')
        i += 1
f.close()
```
