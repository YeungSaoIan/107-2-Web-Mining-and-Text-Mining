```python
import jieba
import matplotlib as mpl
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
#定义一个空字符串
final = ""
#文件夹位置
filename = r"testnewsrecord.txt"
 
#打开文件夹，读取内容，并进行分词
with open(filename,'r',encoding = 'utf-8') as f:
    for line in f.readlines():
        word = jieba.cut(line)
        for i in word:
            final = final + i +" "
```

```python
word_pic = WordCloud(font_path = r'setofont.ttf',width = 2000,height = 1000).generate(final)
plt.imshow(word_pic)
#take out x y
plt.axis('off')
#save picture
plt.savefig(r'output.png')
```
