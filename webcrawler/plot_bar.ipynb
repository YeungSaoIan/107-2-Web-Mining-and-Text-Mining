```python
tmp = []
for article in data['cln_content']:
    a = str.split(article, sep = ' ')
    tmp.append(a)

test = []

num = 0
for num in range(len(tmp)):
       for insideNum in range(len(tmp[num])):
            if len(tmp[num][insideNum]) ==3 :
               test.append(tmp[num][insideNum])     
```
