```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://www.ptt.cc/bbs/movie/index.html") #將此頁面的HTML GET下來
print(r.text) #印出HTML
```

```python
soup = BeautifulSoup(r.text,"html.parser") #將網頁資料以html.parser
sel = soup.select("div.title a") #取HTML標中的 <div class="title"></div> 中的<a>標籤存入sel
```

```python
for s in sel:
    print(s["href"], s.text) 
```

```python
t = requests.get("https://www.chinatimes.com/chinese/total?chdtv") #將此頁面的HTML GET下來
print(t.text) #印出HTML
```

```python
drink = BeautifulSoup(t.text,"html.parser") #將網頁資料以html.parser
giveme = drink.select("h3.title a") #取HTML標中的 <div class="title"></div> 中的<a>標籤存入giveme
showmetime = drink.select("div.meta-info time")
```

```python
for u in giveme:
    print(u["href"], u.text) 
```

```python
for v in showmetime:
    print(v["datetime"])
```

```python
print(u, v["datetime"])
```

```python
for u in giveme:
    print(u)
    for v in showmetime:
        print(v["datetime"]) 
```
